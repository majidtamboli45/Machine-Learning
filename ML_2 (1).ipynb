```python
# 2. Features and target
features = ['season','yr','mnth','hr','holiday','weekday','workingday','weathersit','temp','atemp','hum','windspeed']
target = 'cnt'
```

```python
# 3. One-hot encoding for categorical features
categorical_cols = ['season','mnth','hr','holiday','weekday','workingday','weathersit']
x = pd.get_dummies(df[features], columns=categorical_cols, drop_first=True)
y = df[target]
```

```python
x
```

```python
y
```

```python
numerical_cols = ['temp','atemp','hum','windspeed']
```

```python
# 4. Standardize numerical feature
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x[numerical_cols] = sc.fit_transform(x[numerical_cols])
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size = 0.8 , random_state = 42)
```

## Ridge 

```python
from sklearn.linear_model import RidgeCV

# 6. Ridge Regression with Cross-Validation
ridge_alphas = [0.001, 0.01, 0.1, 1, 10, 100]
ridge = RidgeCV(alphas=ridge_alphas, cv=5)
ridge.fit(x_train, y_train)
y_pred_ridge = ridge.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
# Evaluate Ridge
print("Ridge Regression")
print("Best alpha:", ridge.alpha_)
print("R2 Score:", r2_score(y_test, y_pred_ridge))
print("MSE:", mean_squared_error(y_test, y_pred_ridge))
print("MAE:", mean_absolute_error(y_test, y_pred_ridge))
print("\n")
```

## Lasso

```python
from sklearn.linear_model import LassoCV

# 7. Lasso Regression with Cross-Validation
lasso_alphas = [0.0001, 0.001, 0.01, 0.1, 1, 10]
lasso = LassoCV(alphas=lasso_alphas, cv=5, max_iter=10000)
lasso.fit(x_train, y_train)
y_pred_lasso = lasso.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Evaluate Lasso
print("Lasso Regression")
print("Best alpha:", lasso.alpha_)
print("R2 Score:", r2_score(y_test, y_pred_lasso))
print("MSE:", mean_squared_error(y_test, y_pred_lasso))
print("MAE:", mean_absolute_error(y_test, y_pred_lasso))
print("\n")
```
