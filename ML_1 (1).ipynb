```python
df1.dropna(inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['type'] = le.fit_transform(df1['type'])
```

```python
df1
```

```python
x = df1.drop('type',axis=1)
y = df1['type']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
ste = StandardScaler()
x_scaled = ste.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
model = PCA(n_components=5)
```

```python
data_pca = model.fit_transform(x_scaled)
```

```python
data_pca
```

```python
model.explained_variance_
```

```python
model.score
```

```python
model.explained_variance_ratio_
```

```python
plt.figure(figsize=(10,8))
plt.scatter(data_pca[:,0] , data_pca[:,1] , c = df1['type'] , cmap = "plasma")
plt.colorbar(label = "Wine Quality")
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(data_pca , y , train_size = 0.8 , random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
```

```python
model1
```

```python
model1.fit(x_train,y_train)
```

```python
y_pred1 = model1.predict(x_test)
```

```python
y_pred1
```

```python
from sklearn.metrics import accuracy_score , classification_report
acc = accuracy_score(y_test,y_pred1)
cr = classification_report(y_test,y_pred1)
print(cr)
print("Accuarcy Score is : " , acc)
```
